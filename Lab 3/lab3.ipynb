```python
# Lab 3: Exploratory Data Analysis (EDA)
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()

# Load Dataset
df = pd.read_csv("daily_internet_usage_by_age_group.csv")

# Display first 5 rows
df.head()
```

```python
# Check all values to Check Missing Values
print(df.isna())
```

```python
# HOW many missing vlaues
print(df.isna().sum())
```

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
# viewing the data types of columns
df.dtypes
```

```python
# Average screen time by age group
df.groupby('age_group')['total_screen_time'].mean()
```

```python
# Average work or study time by age group
df.groupby('age_group')['work_or_study_hours'].mean()
```

```python
# Average social media time by age group
df.groupby('age_group')['social_media_hours'].mean()
```

```python
# number of each internet type
df['internet_type'].value_counts()
```

```python
# Statistical summary
df.describe(include='all')
```

```python
# figure of how many people use social media for X hours
sns.histplot(df['social_media_hours'], bins=10)
plt.title("Distribution of Social Media Usage")
```

```python
# figure of how many people have a total screen time for X hours
plt.figure(figsize=(8,5))
sns.histplot(df['total_screen_time'], bins=10)
plt.title("Distribution of Total Screen Time")
plt.xlabel("Total Screen Time (hours)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Total study time by age group
study_time_by_age = (
    df.groupby('age_group')['work_or_study_hours']
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))
study_time_by_age.plot(kind='bar')
plt.title("Total Study/Work Hours by Age Group")
plt.ylabel("Total Hours")
plt.xlabel("Age Group")
plt.show()

study_time_by_age
```

```python
# Average screen time by device
avg_screen_by_device = (
    df.groupby('primary_device')['total_screen_time']
      .mean()
)

avg_screen_by_device.plot(kind='bar')
plt.title("Average Screen Time by Primary Device")
plt.ylabel("Average Hours")
plt.show()
```

```python
# Study/Work Hours vs Total Screen Time
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='work_or_study_hours',
    y='total_screen_time',
    data=df
)
plt.title("Study/Work Hours vs Total Screen Time")
plt.xlabel("Study/Work Hours")
plt.ylabel("Total Screen Time (hours)")
plt.show()
```

```python
# Total Screen Time by Internet Type
plt.figure(figsize=(8,5))
sns.stripplot(
    x='internet_type',
    y='total_screen_time',
    data=df,
    jitter=True
)
plt.title("Total Screen Time by Internet Type")
plt.xlabel("Internet Type")
plt.ylabel("Total Screen Time (hours)")
plt.show()
```

```python
# Correlation between all numeric columns
plt.figure(figsize=(8,6))
sns.heatmap(
    df[['age', 'social_media_hours', 'work_or_study_hours', 'entertainment_hours', 'total_screen_time']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Matrix of Screen Time Data")
plt.show()
```
